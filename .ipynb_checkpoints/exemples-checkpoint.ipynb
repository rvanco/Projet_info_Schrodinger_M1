```python
from Schrodinger import *
```

```python
#Here we define different functions to be used as potentials 
V_0 = 5
a = 2
def V (x):
    #square well
    return -V_0/2*(1-np.sign(np.abs(x)-a))

b = 2
def V_2 (x):
    #quadratic well
    return V_0/2*((x/b)**2-1)*(1-np.sign(np.abs(x)-b))

def Lennard_Jones(a_1,a_2):
    return lambda x : - 4*V_0*( (a_1/x)**6 - (a_2/x)**12)
```

```python
resolution(Lennard_Jones(1/1.12*10,1/1.12*10),-V_0,-0.01,5,40,x_c=10,dx=10**-2,eps=10**-7,err=10**-10)
```

```python
resolution(V_2,-V_0,-0.01,x_m,x_M,x_c=0,dx=10**-2,eps=10**-6,err=10**-2)
```

```python
resolution(V,-V_0,0,x_m,x_M,x_c=0,dx=10**-2,eps=10**-7,err=10**-2)
```
